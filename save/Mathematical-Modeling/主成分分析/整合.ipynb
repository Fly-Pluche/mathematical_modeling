### [PCA官方文档](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html#sklearn.decomposition.PCA)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
scaled_data = StandardScaler().fit_transform(data)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(scaled_data)
```

```python
# 主成分（特征向量）
pca.components_
```

```python
# 贡献率
pca.explained_variance_ratio_
```

```python
plt.matshow(pca.components_,cmap='GnBu')

plt.xticks(range(data.shape[1]),
           data.columns, rotation=0, ha='left')
plt.yticks([0,1],['第一主成分','第二主成分'])
plt.colorbar()
plt.xlabel("特征")
plt.ylabel("主成分")
```

### 手写

```python
n, p = data.shape
cor = data.corr()
# 标准化
scale_data = preprocessing.scale(data)
V,E = np.linalg.eig(cor) # V是特征值，E是特征值对应的特征向量
lamb = np.array(sorted(V, reverse=True))
# 贡献率
contribution_rate = lamb / sum(lamb)
# 累计贡献率
cum_contri_rate = np.cumsum(contribution_rate)
# 按照特征值递减对特征向量排序的结果
A = E[:,V.argsort()[::-1]]
m = 2
F = np.zeros((n,m)) 
for i in range(m):
    ai = A[:,i]
    Fi = (ai * scale_data).sum(axis=1)
    F[:,i] = Fi
```
