```python
import pandas as pd
```

```python
data = pd.read_csv('forest.csv', index_col=0)
data.sample(6)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
standar_data = StandardScaler().fit_transform(data)
standar_data = pd.DataFrame(standar_data, index=data.index, columns=data.columns)
standar_data.sample(6)
```

```python
x = standar_data.iloc[:,0]
y = standar_data.iloc[:,1]
z = standar_data.iloc[:,2]
```

```python
import pyecharts.options as opts
from pyecharts.charts import Scatter3D
from pyecharts.globals import CurrentConfig, NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
```

```python
# data = standar_data.copy()
# data = data.rename(columns={'森林覆盖率（%）':'x','林木蓄积量（亿立方米）':'y','草原面积（万公顷）':'z'})
```

```python
data_ = list(zip(x,y,z))
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x, y, z)
```

```python
_3d = (Scatter3D()
             .add("",
                  data=data_,
                  xaxis3d_opts=opts.Axis3DOpts(
                      name='森林覆盖率（%）',
                      type_="value",
                      # textstyle_opts=opts.TextStyleOpts(color="#fff"),
                      ),
                  yaxis3d_opts=opts.Axis3DOpts(
                      name='林木蓄积量（亿立方米）',
                      type_="value",
                      # textstyle_opts=opts.TextStyleOpts(color="#fff"),
                      ),
                  zaxis3d_opts=opts.Axis3DOpts(
                      name='草原面积（万公顷）',
                      type_="value",
                      # textstyle_opts=opts.TextStyleOpts(color="#fff"),
                      )
                 )
             )

_3d.load_javascript()
```

```python
_3d.render_notebook()
```
