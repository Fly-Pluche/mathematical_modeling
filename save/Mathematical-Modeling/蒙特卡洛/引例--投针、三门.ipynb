### 引例--蒲丰投针

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
```

```python
l = 0.52  # 针的长度
a = 1.314  # 区域高度
n = 10000  # 投针次数
m = 0  # 初始化针与边界的相交次数
# 模拟针的中点与最近边的距离
x = random.uniform(0, a/2, n)  # 随机生成n个[0,a/2]内服从均匀分布的浮点数
# 模拟针与边的夹角
phi = random.uniform(0, np.pi, n)  # 随机生成n个[0,Π]内服从均匀分布的浮点数
new_x = []
new_phi = []
for i in range(n):
    if x[i] <= (l/2) * np.sin(phi[i]):  # 针与边相交的条件
        m += 1
        new_x.append(x[i])
        new_phi.append(phi[i])
        
plt.scatter(phi, x, color='b', s=8)
plt.scatter(new_phi, new_x, color='r', s=8)
p = m / n
pi = (2 * l) / (a * p)
print(pi)
```

### 实例--三门问题

```python
n = 10000  # 模拟次数
a = 0  # 不改变主意赢得的次数
b = 0  # 改变主意赢得的次数
for i in range(n):
    x = random.randint(1,4)  # 奖品所在的门号
    y = random.randint(1,4)  # 自己所选的门号
    if x == y:  # 自己选的是奖品所在门
        a += 1  # 不改变则获奖
    else:
        b += 1
print('改变主意赢得的概率：', b/n)
print('不改变主意赢得的概率：', a/n)
```
