```python
import pandas as pd
pd.set_option('display.float_format',lambda x : '%.2f' % x)  # 保留小数点后4位有效数字
```

```python
data = pd.read_excel('cotton.xlsx')
```

```python
data.head()
```

```python
data.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
year = data['年份']
other = data.iloc[:,1:]
```

```python
sns.set(font='SimHei')
plt.figure(figsize=(10,8))
for i in range(6):
    if i == 0:
        sns.lineplot(year,other.iloc[:,i],label=other.iloc[:,i].name,linestyle=':')
    else:
        sns.lineplot(year,other.iloc[:,i],label=other.iloc[:,i].name)
```
