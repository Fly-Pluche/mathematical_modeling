```python
from scipy.optimize import fsolve
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.interpolate import CubicSpline

x = np.linspace(-10, 10, 40)
y = np.sin(x) + np.random.normal(loc=0, scale=0.05, size=x.shape)
xx = np.linspace(-10, 10, 1000)

fig = plt.figure(figsize=(15, 5))
plt.scatter(x, y, c='r', label='Raw Data')
# kind是表示几次插值. 取3时, 与CubicSpline(三次样条插值无边界条件)效果一样
kind = 5
In = interp1d(x, y, kind=kind)
plt.plot(xx, In(xx), c='b', label='Interpolated Curve of {}'.format(kind))
CS = CubicSpline(x, y)
plt.plot(xx, CS(xx), c='orange', label='Cubic Spline')
plt.plot(xx, np.sin(xx), c='g', label='Standard Curve')
plt.ylabel('x')
plt.legend()
plt.tight_layout()
plt.show()
```
